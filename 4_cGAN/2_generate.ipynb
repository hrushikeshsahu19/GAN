```python
from tensorflow.keras.models import load_model
```

```python
generator = load_model("cgan_generator.h5")
```

```python
from numpy.random import randn, randint
from numpy import asarray
```

```python
def generate_latent_points(n, latent_dim=100):
    x_input = randn(latent_dim*n)
    x_input = x_input.reshape(n, latent_dim)
#     labels = randint(0, classes, n)
    return x_input #[x_input, labels]
```

```python
import matplotlib.pyplot as plt
def plot(examples, n):
    plt.figure(figsize=(12,10))
    for i in range(n*n):
        plt.subplot(n, n, i+1)
        plt.axis("off")
        plt.imshow(examples[i, :, :, 0], cmap="gray_r")
    plt.show()
```

```python
latent_points = generate_latent_points(100)
labels = asarray([x for _ in range(10) for x in range(10)])
X = generator.predict([latent_points, labels])
X = (X+1)/2
plot(X, 10)
```
